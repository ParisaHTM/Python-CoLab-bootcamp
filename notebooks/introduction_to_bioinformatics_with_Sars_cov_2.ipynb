### sars-cov-2 genbank

```python
from Bio import SeqIO
from Bio import Entrez
handle = Entrez.efetch (db = "nucleotide", rettype = 'gb', id = "NC_045512.2", retmode = 'text')
#print(handle.read())
recordRef = SeqIO.read(handle, "genbank") # get the genbank file
```

```python
dir(recordRef)
```

```python
recordRef.seq
```

```python
print(recordRef.seq)
```

```python
print(len(recordRef.seq))
```

```python
recordRef.features
```

```python
#record = SeqIO.read("NC_045512.gb", "genbank") # get the genbank file
#dna_seq = record.seq

CDSs = [feature for feature in recordRef.features if feature.type == "CDS"] #Collect coding sequences
len(CDSs)
```

```python
MatPeptides = [feature for feature in recordRef.features if feature.type == "mat_peptide"] #Collect coding sequences
len(MatPeptides)
```

### Where is the spike gene? 

https://www.ncbi.nlm.nih.gov/nuccore/NC_045512

  CDS             21563..25384 <br> 
                   /gene="S" <br> 
                     /locus_tag="GU280_gp02"<br> 
                     /gene_synonym="spike glycoprotein" <br> 
                     /note="structural protein; spike protein" <br> 

Note: 
python sequence starts at zero and is end exclusive

```python
start = 21563
end = 25384
spike_cds = recordRef.seq[(start-1):end] #python sequence starts at zero and is end exclusive
print(len(spike_cds))
spike_cds.translate()
```

```python
(end - start + 1)/3
```

```python
recordRef.seq[(start-1):(start+2)].translate()
```

```python
recordRef.seq[(end-3):end] #stop codon
recordRef.seq[(end-3):end].translate()
```
